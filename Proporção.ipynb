```python
from google.colab import drive
import pandas as pd

drive.mount('/content/drive')

caminho_arquivo = '/content/drive/MyDrive/Colab_Notebooks/dataset/participants.tsv'

df = pd.read_csv(caminho_arquivo, sep='\t')

total_pacientes = len(df)
contagem = df['sex'].value_counts()
proporcao = (contagem/total_pacientes)*100

print(f"Total de pacientes no arquivo: {total_pacientes}\n")
print("--- Quantidade Absoluta ---")
print(contagem.to_string())
print("\n--- Proporção (%) ---")
print(proporcao.round(2).to_string())
```
