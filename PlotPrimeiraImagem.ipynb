```python
import numpy as np
import matplotlib.pyplot as plt

caminho_arquivo = "/content/drive/MyDrive/Colab_Notebooks/dataset/sub-114950860875_preproc-cat12vbm_desc-gm_T1w.npy"

dados_volume = np.load(caminho_arquivo)

# remove as dimensões de tamanho 1
dados_volume = np.squeeze(dados_volume)

print(f"Dimensões corrigidas para plotagem: {dados_volume.shape}")

meio_x = dados_volume.shape[0] // 2
meio_y = dados_volume.shape[1] // 2
meio_z = dados_volume.shape[2] // 2

fatia_sagital = dados_volume[meio_x, :, :]
fatia_coronal = dados_volume[:, meio_y, :]
fatia_axial   = dados_volume[:, :, meio_z]

fig, axes = plt.subplots(1, 3, figsize=(15, 5))

axes[0].imshow(np.rot90(fatia_sagital), cmap='gray')
axes[0].set_title('Corte Sagital')
axes[0].axis('off')

axes[1].imshow(np.rot90(fatia_coronal), cmap='gray')
axes[1].set_title('Corte Coronal')
axes[1].axis('off')

axes[2].imshow(np.rot90(fatia_axial), cmap='gray')
axes[2].set_title('Corte Axial')
axes[2].axis('off')

plt.show()
```
